__Section 1: Data and Research Questions__

_Data Set_

This study is taken from Rodgers et al. 2022: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC9118540/
In this paper the authors investigate the prevalence of PFAS and PFAS analytes in consumer
items used by children and adolescents, and the reliability of green assurances on product labels relative
to the product’s PFAS level. To do this the authors gathered the following information from 93 products
marketed towards children and adolescents:
-  What type of product it was: furnishings, apparel, or bedding
-  What type of “green language” and/or “water resistant” claims were made on the product
packaging
-  Where the product was produced
- Each product’s fluorine level and PFAS/PFAS analytes level

Therefore, the observations of this study are the 93 products tested, and the features of each
product are the packaging labels, the product’s place of origin, and toxicity levels in regards to fluorine
and PFAS/PFAS analytes. These features would tell us whether consumers can rely on “green assurances”
used on packaging in order to avoid exposure to toxic PFAS. These features would also tell us if
consumers should avoid “water/stain resistant” products when trying to avoid PFAS in their products. If
certain countries are producing higher levels of PFAS in their products, consumers could also use the
location of origin of their products as a factor when trying to avoid PFAS. In summary, these features
would tell us what parts of product marketing can and cannot be used when trying to avoid toxic
chemicals.

_Research Questions_

The research question of the authors was: is the language used on product labels indicative of
PFAS toxicity in said product? Researchers tested fluorine and PFAS levels in each product and used that
data to see if there was a difference of PFAS presence between product labels that were trademarked, used
green language, and/or claimed to be water resistant and those that were not. This study focused on
products used by children and teens because research has shown that children and adolescents are
particularly vulnerable to the negative health impacts that PFAS can have on the human body.

Another research question I could ask using this data would be whether the products that use
“green” language on their product labels have a lower total fluorine level. It is possible that the products claiming to be the
cleanest actually have the same fluorine level as those that do not have the same assurances. Total fluorine detected in samples was used as delimiting factor for PFAS testing. Samples that tested below 10 ppm were believed to not have a significant amount of PFAS analytes and, therefore, were not analyzed for PFAS analytes. If a relationship exists between the presence of green assurances on a product label and the total fluorine of a product, then the marketing language used product labels could be an indicator of PFAS presence. This analysis would further support the researcher’s
hypothesis of the lack reliability of product labels in regards to product toxicity if the products with the green
marking turn out to be the same, in terms of fluorine level, as those that do not.

_Visualization_

In order to compare these two features, we must first group the label groups by whether or not they used green assurances. Product samples from label groups 'WS-G' and 'G' will be grouped together as the 'green marketing language' group. Product samples from label groups 'WS-T', 'WS', and 'N-WSG' will be grouped together as the 'no green marketing language used' group. Then, for all samples with a recorded value for label group that passed the total fluorine delimiting factor, we must match the product sample's total fluorine level with either the 'green language used' or 'green language not used' marketing language group. 

```python
#Create two variables 'has_green' and 'no_green'
    #in 'has_green' store the list containing the label groups that use green assurances
    #in 'no_green' store the list containing the label groups that do not use green assurances
has_green = ['WS-G', 'G']
no_green = ['WS-T', 'WS', 'N-WSG']

#Create an empty dictionary that can be used to store the values for total fluorine and green language group
fluorine_and_green = {}

#Create an empty list for the green label groups
green_label = []

#Create a variable called 'fluorine_label' that contains the values for total fluorine 
#and label group 
    #This variable is only for samples that passed the total fluorine delimiting factor and have a recorded value for label group
    #This is 65 samples
fluorine_label = PFAS_products.loc[PFAS_pass_delimiter,['Total F (ppm)', 'Label groupa']].dropna()

#for all values in the column 'Label groupa' in 'fluorine_label'
    #if the value is in 'has_green'
        #add 1 to the end of the 'green_label' list
    #if the value is in 'no_green'
        #add 0 to the end of the 'green_label' list
for i in fluorine_label['Label groupa']:
    if i in has_green:
        green_label.append(1)
    if i in no_green:
        green_label.append(0)


#Add the total fluorine values from 'fluorine_label' to the dictionary under the key 'Total F (ppm)'
    #convert these values to integer form
#Add the list 'green_label' to the dictionary under the key 'Green Label'
fluorine_and_green['Total F (ppm)'] = fluorine_label['Total F (ppm)'].astype(int)
fluorine_and_green['Green Label'] = green_label

#Use Pandas to turn the dictionary into a data frame
fluorine_and_green_df = pd.DataFrame(fluorine_and_green)

#Print the data frame
fluorine_and_green_df
```

The type of visualization I chose to make was a distribution of the total fluorine level broken up by green label group.
```python
#Create a distribution plot of total fluorine for both the green and not green label group
    #set the data as the new data frame 'fluorine_and_green_df'
    #seperate the distributions by whether green language was used on the label
        #with 1 meaning green language was used
        #with 0 meaning green language was not used
    #set the 'common_norm' to false so the distribtuions are drawn independently of each other
    #set the stat to density to plot the probability of the distribution within the data set
    
sns.displot(data=fluorine_and_green_df, 
            x="Total F (ppm)", 
            col='Green Label',
            stat='density',
            common_norm=False, 
            kde=True)
```

These plots are useful in determining whether products marketed as green actually contained less fluorine. As shown by these plots, there is almost no clear difference found in the distribution of total fluorine between the green and not green groups. _Model_

The target feature we are modeling is the numeric feature: total fluorine level of product samples that passed the delimiting factor. This is the dependant feature. The target feature will by modeled by the binary feature: whether the product, that the sample was taken from, label contained green assurances or not. This is the independant feature.

A linear regression model is appropriate as it will show the effect, if any, that the presence of green assurances have on the total fluorine level of a product sample. If no relationship is found between these two features, then my own hypothesis and the researcher's hypothesis that marketing language is not a good indicator of product toxicity will be supported. 

```python
#In this line we add a column of all 1's to our 'fluorine_and_green_df' dataframe, 
#so that sm.OLS().fit() can find our intercept

fluorine_and_green_df = sm.add_constant(fluorine_and_green_df)

#Create and fit the model, and print out a summary
    #set the variable x as the independant variable containing 'const' and 'Green Label' from the dataframe
    #set y as the depedant variable 'Total F (ppm)' from the data frame
    #Model total fluorine in terms of green label group

X = fluorine_and_green_df.loc[:,['const','Green Label']]
y = fluorine_and_green_df['Total F (ppm)']
model_fluorine_and_green = sm.OLS(y, X).fit()
model_fluorine_and_green.summary()
```

```python
model_fluorine_and_green.params
```

The 'const' value in params is the same as the y-intercept for this model, or the predicted average total fluorine for products that did not have green language on their label. The 'Green Label' value in params is the slope. This is how much the presence of green language in product label effects the total fluorine level of a product, meaning the presence of green language decreases the average total fluorine present in a sample by about 81 ppm. 

The standard error for the slope is very large, meaning that this model does not fit the observed data well. The confidence interval ranges from negative to positive. The p-value, at 0.691, is very large meaning that green marketing language is not a good predictor of total fluorine. Thus, from these three factors, we can say that there is no relationship between the presence of green assurances on a product's label and total fluorine level detected in the product sample. We cannot be certain of this model.

In terms of fluorine level, products that are being marketed as green are no different from the products that are not marketed as green. Therefore, consumers cannot rely on product labels when trying to stay away from products containing fluorine. My hypothesis is supported. This supports the researcher's hypothesis that a product's label is not a good indicator of the presence of fluorine, and therefore possibly PFAS, in products marketed to children and teenagers.

```python
#Create an empty list for the green label groups
    #with 1 meaning green language was used
    #with 0 meaning green language was not used
mass_green = []


#for all values in the column 'Label groupa' in 'pass_delimiter_label_country'
    #if the value is 'WS-T'
        #add 0 to the end of the 'country_green_groups' list
    #if the value is 'WS'
        #add 0 to the end of the 'country_green_groups' list
    #if the value is 'WS-G'
        #add 1 to the end of the 'country_green_groups' list
    #if the value is 'G'
        #add 1 to the end of the 'country_green_groups' list
    #if the value is 'N-WSG'
        #add 0 to the end of the 'country_green_groups' list
for i in label_mass['Label groupa']:
    if i == 'WS-T':
        mass_green.append(0)
    if i == 'WS':
        mass_green.append(0)
    if i == 'WS-G':
        mass_green.append(1)
    if i == 'G':
        mass_green.append(1)
    if i == 'N-WSG':
        mass_green.append(0)

#Create an empty dictionary that can be used to store the values for sample mass per area (mg/cm2) and green label groups
mass_and_green = {}

#Under the key 'Sample mass per area (mg/cm2)', add the integer values of the 'Sample mass per area (mg/cm2)' column from the data from 'label_mass'
#Under the key 'Green Label', add the list 'mass_green'
mass_and_green['Green Label'] = mass_green
mass_and_green['Sample mass per area (mg/cm2)'] = label_mass['Sample mass per area (mg/cm2)'].astype(int)

#Use Pandas to turn the dictionary into a data frame
mass_and_green_df = pd.DataFrame(mass_and_green)

#In this line we add a column of all 1's to our 'Country_and_label' dataframe, 
#so that sm.OLS().fit() can find our intercept
mass_and_green_df = sm.add_constant(mass_and_green_df)

#Create and fit the model, and print out a summary
x_mass_green = mass_and_green_df.loc[:,['const','Green Label']]
y_mass_green = mass_and_green_df['Sample mass per area (mg/cm2)']
mass_and_green_model = sm.OLS(y_mass_green, x_mass_green).fit()
mass_and_green_model.summary()
```
